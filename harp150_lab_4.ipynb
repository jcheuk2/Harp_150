```python
# Write the code that expresses these conditionals. Don’t forget the difference between = and == !
# What is the weather today?

# If the weather is sunny→ Wear sunglasses.
# Or if the weather is rainy → Bring an umbrella.
# Or if the weather is cold → Wear a sweater.
# Or if the weather is hot → Wear shorts.
# Otherwise if the weather is none of the above → Check the weather channel.

weather = input("what is the weather like today? ")

if weather == "sunny":
  print("Wear sunglasses")
elif weather == "rainy":
    print("bring an umbrella")
elif weather == "cold":
  print("wear a sweater")
elif weather == "hot":
  print("wear shorts")
else:
  print("check the weather channel")
```
