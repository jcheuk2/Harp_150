***Deleting columns***


* Let’s say we don’t need the CommunityBoard(s) column for our data analysis. We can delete it using .drop() method, so that the dataset is easier to work with

* (FYI: If you want to drop multiple columns, you just put them in as a list)
* (Note that it is best practice to avoid using spaces in your column names. Otherwise, it may error out in some situations, like when you try to access a column using dot notation, like permits.EventID - more on it in the next slide)



```python
permits1=permits.copy()

permits1.drop("CommunityBoard(s)", axis = 1, inplace = True)
# inplace = true means that we want to drop() to modift the df in place/make permanent change
# axis = 1 means columns, you can also use axis = "columns"
# similarly, axis = 0 means rows. if you do not specify acis, the default is 0

# alternatively, we can use this to achieve the same result
# permits1.drop(columns = "CommunityBoard(s)", inplace = True)

permits1
```

*Do a simple data analysis *

* Let’s say we want to count how many permits were given out in each borough.
* We can use a for loop & a dictionary to figure that out pretty quickly
* But pandas also has a built-in method to accomplish the same goal: .value_counts()
* We will get the same values from both solutions


```python
permit_dict1 = {}
for i in permits ["Borough"]:
  if i not in permit_dict1:
    permit_dict1[i] = 1
  else:
    permit_dict1[i] += 1

# using built in pandas method
permit_result = permits["Borough"].value_counts()

print(permit_dict1)
print(permit_result)
```

**Exporting your DataFrame to .CSV**
* Lastly, we want to export our little analysis into .csv, so we can use it elsewhere →  df.to_csv(“path/name.csv”) or df.to_csv(“name.csv”)
* We can use this method directly on permit_result, since pandas built-in method .value_counts() outputs a df or a series directly for us

* However, since permit_dict1 is a python dictionary, we need to turn the dictionary into a DataFrame FIRST by using pd.DataFrame(), a pandas function to create df objects
* There are two ways to output/structure the new df, which will result in two different shapes of the same data:
a wide/unstacked shaped df & a long/stacked shaped one (see the next slide)
* Both shapes are valid, but depending how we want to use the dataset, we may prefer one shape over the other.
* See more examples on creating df [here](https://www.geeksforgeeks.org/how-to-create-dataframe-from-dictionary-in-python-pandas/)


You would want to expirt dataframe to .csv becasue you could use it

```python
permit_df1 = pd.DataFrame(permit_dict1, index = [0])
permit_df2 = pd.DataFrame(permit_dict1.items(),columns=["BoroughName","PermitCounts"])
permit_df2.to_csv("sample_data/film_permits_results2.csv")
# csv = Comma-separated values


print(permit_df1)
print(permit_df2)
```
