What is a while loop? when should you use them?

A for loop is used for a specific length of items to loop through

a while loop is used to repeat a block of code until a condition is met

While loops are used when you do not have a preexisting finite set of information to loop through (like a range, list, or dictionary)
They will always have a condition and the code indented after it will be run while that condition is true and stop when it becomes false.
Sometimes the condition is never false and the loop will run forever! That is called an â€œinfinite loopâ€ and you rarely want one.


```python
# example let's go through this line by line on the board:

x = 1
total = 0

while x <= 10:
  total += x #this means that x +total = total. the new total is x + total
  x += 1 #x + 1 = X . this means that x is x + 1, because there's a new value of x every time,
  print(x)
print(total)

# if you don't know what math it's doing,
#tell the computer to print the variable to see the numbers
#to check how x and total are changing, ask computer to print them
```
