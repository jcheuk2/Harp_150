# Turn a series into a dataframe and set new column names


* Previously, we have learned using for loops and built-in method .value_counts(“column_name”) to do counting by column values.
* Let’s use .value_counts() to find out how many homes are listed per year:


* The method actually returns us a series, a pandas data structure. A series is one-dimensional array, that is a column or a row. In comparison, a dataframe is two-dimensional array, that is a collections of columns.


* Next, we want to turn our returned pandas series into a df and give it associated column names “Year” and “Count” !
(Note: you may wonder since the definition of a series is that it is only one column or a row. Then, how come we have two columns of data, i e. years and counts, in our series? The thing is years is actually used as the index of the series, which does not count as a column.)


* This also explains why .reset_index() actually makes a series into a df. By adding index, the original index years become a real column. Thus, it turned a series from one dimensional to two, which is a dataframe!)
pd.DataFrame() is also a common way to form a df!



````python
# you're adding a new dataframe

yearly_counts = ct_df["List Year"].value_counts()

yearly_counts

# below are series, a one dimensional array/ long list of data
````

````python
# two ways to ocnvert a series into a data frame (DF)
# using reset_index(). this makes a series into a data frame

yearly_counts_df1 = yearly_counts.reset_index()
yearly_counts_df1.columns = ["Year","Count"]

# using pd.dataframe
yearly_counts_df2 = pd.DataFrame({"Year": yearly_counts.index, "Count":yearly_counts.values})

yearly_counts_df2

# this kinda pairs the data together like a dictionary, so the year correspons to the cout
````

# find mean

* We’re ready to find the mean (average) house price per year, we can accomplish this with .groupby().
* This takes a column (in this case “List Year”) and groups it with “Sale * * * Amount”, which we then combining them to get the average using mean()
* We can also use common aggregation functions like sum(), max(), min() for other scenarios

* Notice that I am storing this new data frame in the variable avg_sale and using the .round() to make the numbers more readable
* You can understand groupby() as “group column A by column B”; [[]] ensures the result being a df, [] will result in a series.



````python
avg_sale = new_df.groupby("List Year")[["Sale Amount"]].mean().round()

avg_sale
````
