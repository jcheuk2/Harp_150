````python
global_var = "I am a global variable "

def check_scope():
  # local scope
  local_var = "i am a local variable"
  print(global_var)
  print(local_var)

check_scope()

print(local_var)
````

````python
global_var = "I am a global variable "

def check_scope():
  # local scope
  global local_var
  local_var = "i am a local variable"
  print(global_var)
  print(local_var)

check_scope()

print(local_var)
````

3. OOP : class attributes vs instance attributes

````python
class Book:
  medium = "paperback"
  # medium here is the class attribute
  def __init__(self, author, title):
    #instance attributes
    self.author = author
    self.title = title
# creating instances
book1 = Book("jane austen", "pride and prejudice")
book2 = Book("Lewis Carroll", "alice in wonderland")

print(Book.medium)
print(book1.medium, book1.author, book1.title)
print(book2.medium, book2.author, book2.title)

````

5. Error handling: try & except ( [all the built-in exception errors](https://www.w3schools.com/python/python_ref_exceptions.asp))


````python
def divide_numbers(num1,num2):
  try:
    result = num1/num2
    print(f"The result is {result}")
  except ZeroDivisionError:
    print("you can't divide by zero!")
  except TypeError:
    print("please provide numbers only!")
divide_numbers(10,2)
divide_numbers(10,0)
divide_numbers("10",2)
````

Why do we need this over conditionals?
