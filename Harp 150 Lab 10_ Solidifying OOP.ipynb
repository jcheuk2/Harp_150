***Another principle of OOP: Polymorphism***

* I’ve already seen some of you running into polymorphism in your game projects, so let’s explore this →
* Polymorphism allows something to take multiple forms. The behavior of an object or function is based on the context or input. That means, your program will determine which meaning/usage is necessary for each individual executive of an object.
* “Polymorphism” literally means many forms – it’s just a fancy word to say something can do many different things
* Real-life example: your driver’s license both conveys your ability to legally drive a car,  and serves as proof of your legal identity



You’ve already been using polymorphism when you use the + operator!

It can: (1) add integers together and (2) concatenate strings

The computer uses context to know which to implement






** Common ways of Polymorphism:
Method Overriding **

* You can create a child class, which has a method that is already defined in its parent class. The child class method will override the parent class method that shares the same name.
* Often, when you want to use method overriding, it is a good practice to have the same method name, same numbers and types of parameters, and the same return type.
* See how both the Turtle and the Bunny class have the same methods? But they run differently because Python knows to differentiate based on context.

***Common Program/Code Structure***

To help with readability, modularity, scalability of code, plus as a convention and a general good practice, we organize our code in the following way:




1.   Import statements always come first
2. Driver/ main function that controls the program’s overall flow (such as main loop, object creation, method/function calling)
3. Independent functions (functions that do not belong to any classes)
4. Classes and their methods
5. Call your driver/ main function at the very end
6. Use comments and documentations to guide people
7. Incorporate error handling
