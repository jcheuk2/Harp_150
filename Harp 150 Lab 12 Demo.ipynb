## How much have house prices increased, on average, between 2010 and 2021 in Connecticut?

```python
#making a copy in case we do something wrong, we can always
#roll the original dataset back

ct_df2 = ct_df1.copy()
```

```python
#filtering the data, we want the years from >= 2010
#filtering conditions are always a boolean (true/false)

filtered_df1 = ct_df2[ct_df2["List Year"] >= 2010]
```

```python
##using the drop() method
#we need index numbers to drop rows of data
#column labels/indexs to drop columns of data
# .index accesses the index labels of the rows in the df

filtered_df2 = ct_df2.drop(ct_df2[ct_df2["List Year"] < 2010].index)
```

```python
#iloc - row and column indexes
#loc - row and column labels

filter_df3 = ct_df2.iloc[(ct_df2["List Year"] >= 2010).values]
filtered_df4 = ct_df2.loc[ct_df2["List Year"] >= 2010]
```
