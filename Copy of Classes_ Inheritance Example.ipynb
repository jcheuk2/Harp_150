```python
#created by Jenn Seibert

class Course:
  def __init__(self, enrolled, capacity, label):
    self.enrolled = enrolled
    self.capacity = capacity
    self.label = label

  def seats_left(self):
    return self.capacity - self.enrolled

  def add_description(self, description):
    self.description = description
    print(self.description)

  def describe_course(self):
    # methods can call other methods of the class by using self. in front of the name 
    print(self.label, " course description is ", self.description, "and there are ", self.seats_left(), "seats available!")
  

class Coding(Course):
  def __init__(self, enrolled, capacity, label, topics):
    super().__init__(enrolled, capacity, label)
    self.topics = topics

  def add_topic(self, item):
    self.topics.append(item)
    print(self.topics)

  def good_room(self, has_projector, room_num):
    # if has_projector is False, this would be a bad room for a coding class; if True, lets assign room num to be a attribute for coding course
    if has_projector == False:
      return # do nothing in this case, we don't want this room!
    else:
      self.room_num = room_num
      print("This course will be in", self.room_num)

  
# a basic course example 
hws215 = Course(12, 25, "HWS 215")
hws215.add_description("Weight lifting -- counts for wellness gen ed")
print(hws215.seats_left())
print("----------------")

# make a coding course
harp150 = Coding(12, 25, "HARP150", ["variables"])
print(harp150.seats_left())
harp150.add_description("A coding course for Harpur students!")
harp150.add_topic("loops")
harp150.good_room(True, "CW 110")
print("----------------")

#make another coding course
harp151 = Coding(24, 25, "HARP151", ["apis"])
harp151.add_description("Build upon Harp 150 skills and learn more Python!")


# you can build a list out of objects!
print("-----------------")
coding_courses = [harp150, harp151] # use the instance variable names as members of your list.

# we can loop through the list and call methods on those objects 
# if we wanted a brief overview of all the coding courses when registering for classes, we can build more Coding Course objects and add them to our list!
for course in coding_courses:
  course.describe_course()
```
