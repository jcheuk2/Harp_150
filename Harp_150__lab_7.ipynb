**Object Oriented Programming**

As we start to build more complex programs, OOP lets us design our programs around objects (or data), objects can contain a number of methods (functions) and specific attributes
* Lucky for you, we’ve been working with classes and objects all along!
Everything you declare in Python is an object, and it has a class.

**Principles of OOP #1: Encapsulation**

Encapsulation is when all the parts and tools you need get packaged together—encapsulated in a class, whose individual instances are known as objects.
* Another way to understand this: a class is a blueprint of an object!
* Other objects do not have access to this class or the authority to make changes, but are only able to call a list of public functions or methods
* This allows us to focus on one part of a program at a time

Every class starts with an __init__ method.
“init” is short for “initialize.”
* This is where we define the class’ variables (properties)
* It tells you that “Every time you make an object from this class, do what’s in here.”
* For this reason, it is known as a constructor
* Why do you need this? Python requires it
* If you don't provide an __init__ method, Python will use a default one. You can still create instances of the class, but those instances will simply not have any attributes associated with the class when you create them.

You will always use self in your method, self is just a stand-in for each object that we will make from this class.
