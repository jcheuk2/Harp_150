How to normalize dataset in a way where you can clean the major dictionary

How to use substrings that can group lots of data together

* A simple way to clean our major dictionary is to incorporate substrings (a substring can be any part of a string) - that's what we're doing here
* (a more complex but more powerful way is to use regular expressions)
* byt we swill stick to what we have learned in python already

* our dat acleaning method is very aggressive, but it gets us somewhere and is based on what we learned so far. We can clean it further if we want
* look at our code here: in the process of dat acleaning, we made decisions to get more normalized data, but we may also havel ost some data

```python
normalized_majors = {}

for major in df.Major:
  try:
    strings = major.lower().strip()
    if "undecided" in strings:
      major = "Undeclared"
    if "eco" in strings:
      major = "Economics"
    if "math" in strings:
      major = "Mathematics"
    if "poli" in strings:
      major = "Political Science"
    if "bio" in strings:
      major = "Biology"
    if 'psyc' in strings:
      major = 'Psychology'
    if 'engi' in strings:
      major = "engineering"
    else:
      major = major.title()
  finally:
    if major not in normalized_majors:
      normalized_majors[major]=1
    else:
      normalized_majors[major] += 1

normalized_majors
```

```python
normalized_states = {}

for state in df.State:
  try:
    strings = state.lower().strip() #.lower() makes all the strings in the dictionary lower;
    #.strip() removes all leading or trailing whitespace
    if "ny" in strings:
      state = "New York"
    if "New york" in strings:
      state = "New York"
    if "NYS" in strings:
      state = "New York"
    if 'i think NY?':
      state = "New York"
    if 'New York, descended of Italian and Brazilian immigrants' in strings:
      state = "New York"
    if 'NY' in strings:
      state = "New York"
    if "Not from US" in strings:
      state = "Not From the US"
    if 'new york' in strings:
      state = "New York"
    else:
      state = state.title()
  finally:
    if state not in normalized_states:
      normalized_states[state] = 1
    else:
      normalized_states[state] += 1


normalized_states
```
